```python
def clean(df, coluna, tratado, bTrata, novo):
    if not bTrata:
        df = df[df[coluna] != novo]
    else:
        df[coluna] = df[coluna].replace(tratado, novo)
    return df
```

```python
df_fighter_data = df_fighter_data.dropna(subset=['STANCE'])

def transformaEmCm(altura):
    if altura != "--":
        if len(altura) > 3:
            cm = float(altura[0]) * 30.48 + float(altura[2:len(altura)-1]) * 2.54
        else:
            cm = float(altura[:len(altura)-1]) * 2.54
        return cm
    return None
    
    
def transformaEmKg(lbs):
    if lbs != "--":
        kg = float(lbs[:len(lbs)-4]) * 0.453592
        return kg
    else:
        return lbs
    
def transformaData(data):
    if data != "--":
        return datetime.strptime(data, "%b%d,%Y").strftime('%d/%m/%Y')
    return data

def retornaIdade(data):
    if data != "--":
        
        dataFormatada = datetime.strptime(data, "%b%d,%Y")

        hoje = datetime.today()

        idade = hoje.year - dataFormatada.year - ((hoje.month, hoje.day) < (dataFormatada.month, dataFormatada.day))

        return idade
    return None

def limpaPercentagem(porcentagem):
    if len(porcentagem) > 2:
        return float(porcentagem[:2])/100
    else:
        return float(porcentagem[0])/100




df_fighter_data.loc[:, "Height_cm"] = df_fighter_data["Height"].apply(transformaEmCm)
mediana_alturas = df_fighter_data["Height_cm"].median()
df_fighter_data["Height_cm"].fillna(mediana_alturas, inplace=True)

df_fighter_data.loc[:, "Age"] = df_fighter_data["DOB"].apply(retornaIdade)
mediana_alturas = df_fighter_data["Age"].median()
df_fighter_data["Age"].fillna(mediana_alturas, inplace=True)



df_fighter_data.loc[:, "Reach_cm"] = df_fighter_data["Reach"].apply(transformaEmCm)
df_fighter_data.loc[:, "Weight_kg"] = df_fighter_data["Weight"].apply(transformaEmKg)
df_fighter_data.loc[:, "DOB_p"] = df_fighter_data["DOB"].apply(transformaData)
df_fighter_data.loc[:, "TDAcc.Sp"] = df_fighter_data["TDAcc."].apply(limpaPercentagem)
df_fighter_data.loc[:, "TDDef.Sp"] = df_fighter_data["TDDef."].apply(limpaPercentagem)
df_fighter_data.loc[:, "Str.Acc.Sp"] = df_fighter_data["Str.Acc."].apply(limpaPercentagem)
df_fighter_data.loc[:, "Str.DefSp"] = df_fighter_data["Str.Def"].apply(limpaPercentagem)



le = LabelEncoder()
df_fighter_data.loc[:, "STANCE_encoded"] = le.fit_transform(df_fighter_data["STANCE"])

df_fighter_data = df_fighter_data.dropna(subset=['STANCE'])

pd.set_option('display.max_rows', None) #Showing only two columns
pd.set_option('display.max_columns', None) #Showing only two columns

df_fighter_data
```
